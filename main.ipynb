# Datasets

```python
import numpy as np
import pandas as pd

df = pd.read_csv("./data/diabetes.csv")
df.head()
```

```python
df.shape
```

```python
df["Outcome"].value_counts()
```

```python
x = df.drop("Outcome", axis=1)
y = df["Outcome"]
```

```python
x
```

```python
y
```

```python
x = np.array(x)
y = np.array(y)
```

```python
x
```

```python
y
```

# Normalize

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
x
```

# Train/Test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

# ANN

```python
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(hidden_layer_sizes=128, max_iter=200)
model.fit(x_train, y_train)
```

```python
y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score

acc_train = accuracy_score(y_train, y_pred_train)
acc_test = accuracy_score(y_test, y_pred_test)

acc_train, acc_test
```

```python
p = precision_score(y_test, y_pred_test)
r = precision_score(y_test, y_pred_test)

p, r
```

# PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
pca.fit(x_train)

x_new_train = pca.transform(x_train)
x_new_test = pca.transform(x_test)
```

```python
x_train.shape, x_new_train.shape
```
