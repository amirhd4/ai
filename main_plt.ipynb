```python
from cProfile import label

import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5,6, 7 ,8, 9]
y = [2, 4, 3, 7, 8, 1, 3, 9, 11]

plt.plot(x, y, marker="o", color="green", label="Matplotlib")
plt.title("Example")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x = ["red", "green", "blue", "orange"]
y = [5, 19, 2, 4]
plt.bar(x, y, color=["red", "green", "blue", "orange"])
plt.title("Example")
plt.xlabel("Color")
plt.ylabel("Count")
plt.show()
```

```python
x = ["red", "green", "blue", "orange"]
y = [5, 19, 2, 4]

plt.pie(y, labels=x, autopct="%1.1f%%", colors=["red", "green", "blue", "orange"])
plt.title("Example")
plt.show()
```
